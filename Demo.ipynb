## Data Analysis

*only 3~5 books of 10 category for the demo*

```python
book_data_file_name = "book_data_for_demo.csv" # *only 3~5 books of 10 category for the demo*
df_book = pd.read_csv(book_data_file_name)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

### Star

```python
df_book_star = df_book.groupby(df_book["main_category"], as_index=True).agg({'aladin_star':'mean',
                                                   'kyobo_star':'mean',
                                                   'yes24_star':'mean',
                                                   'interpark_star':'mean'}
                                                    )
df_book_star
```

```python
mpl.rc('font', family='AppleGothic') #맥 한글 설정
gf_star = df_book_star.plot(kind="bar", figsize=(20,10), fontsize=20)
gf_star.set_title("Star", fontdict={"fontsize":35})
gf_star.set_ylabel("Star", fontdict={"fontsize":30})
gf_star.set_xlabel("Main Category", fontdict={"fontsize":30})
```

### Price

```python
df_book_price = df_book.groupby(df_book["main_category"], as_index=True).agg({'aladin_price':'mean',
                                                   'kyobo_price':'mean',
                                                   'yes24_price':'mean',
                                                   'interpark_price':'mean'}
                                                    )
df_book_price
```

```python
mpl.rc('font', family='AppleGothic') #맥 한글 설정
gf_price = df_book_price.plot(kind="bar", figsize=(20,10), fontsize=20)
gf_price.set_title("Price", fontdict={"fontsize":35})
gf_price.set_ylabel("Price", fontdict={"fontsize":30})
gf_price.set_xlabel("Main Category", fontdict={"fontsize":30})
```

### Review

```python
df_book_review = df_book.groupby(df_book["main_category"], as_index=True).agg({'aladin_review':'mean',
                                                   'kyobo_review':'mean',
                                                   'yes24_review':'mean',
                                                   'interpark_review':'mean'}
                                                    )
df_book_review
```

```python
mpl.rc('font', family='AppleGothic') #맥 한글 설정
gf_review = df_book_review.plot(kind="bar", figsize=(20,10), fontsize=20)
gf_review.set_title("Review", fontdict={"fontsize":35})
gf_review.set_ylabel("Review", fontdict={"fontsize":30})
gf_review.set_xlabel("Main Category", fontdict={"fontsize":30})
```
